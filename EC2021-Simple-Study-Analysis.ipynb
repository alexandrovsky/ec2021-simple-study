```python
import pandas as pd
import re
```

```python
df = pd.read_csv("EC-2021-Unity-Simple-Game-Log - Formularantworten 1.csv")

df.head()
```

```python
pens_cols = df.filter(regex='Reflect on your play experiences and rate your agreement with the following statements:').columns.values
pens_cols

pens_cols_dict = {}

for col in pens_cols:

    q = col[col.find("[")+1:col.find("]")] #re.findall('\[.*?\]',col)
    #print(q[0])
    pens_cols_dict[col] = q

print(pens_cols_dict)
df.rename(columns=pens_cols_dict, inplace=True)
```

```python
pens_cols_dict.values()
```

```python
pens_subscales_dict = {
    'I feel competent at the game' : 'comp_1',
    'I feel very capable and effective when playing' : 'comp_2', 
    "My ability to play the game is well matched with the game's challenges": 'comp_3', 
    'The game provides me with interesting options and choices': 'aut_1', 
    'The game lets you do interesting things' :'aut_2', 
    'I experienced a lot of freedom in the game':'aut_3',  
    'I find the relationships I form in this game fulfilling':'rel_1',  
    'I find the relationships I form in this game important':'rel_2',  
    'I don’t feel close to other players':'rel_3_R',  
    'When playing the game, I feel transported to another time and place': 'pres_1', 
    'Exploring the game world feels like taking an actual trip to a new place': 'pres_2',  
    'When moving through the game world I feel as if I am actually there': 'pres_3',  
    'I am not impacted emotionally by events in the game': 'pres_4_R',  
    'The game was emotionally engaging': 'pres_5',  
    'I experience feelings as deeply in the game as I have in real life': 'pres_6',  
    'When playing the game I feel as if I was part of the story': 'pres_7',  
    'When I accomplished something in the game I experienced genuine pride': 'pres_8',  
    'I had reactions to events and characters in the game as if they were real': 'pres_9',  
    'Learning the game controls was easy': 'ctrl_1', 
    'The game controls are intuitive': 'ctrl_2',
    'When I wanted to do something in the game, it was easy to remember the corresponding control': 'ctrl_3'
}


df.rename(columns=pens_subscales_dict, inplace=True)
```

```python
## Encode values as numerical
```

```python
pens_values_dict = {
    'strongly disagree' : 1,
    'moderately disagree': 2,
    'slightly disagree': 3,
    'neutral': 4,
    'strongly agree' : 5,
    'moderately agree': 6,
    'slightly agree': 7,
}


df.replace(pens_values_dict, inplace=True)
```

```python
## Recode reveres variables
```

```python
reverse_cols = df.filter(regex='_R').columns.values
cols = []
for rcol in reverse_cols:
    col = rcol.replace('_R', '')
    cols.append(col)
    df[col] = 8 - df[rcol]
#df[cols]
cols.extend(reverse_cols)
df[cols]
```

```python
## calculate the subscale scores


```

```python
df['Competence'] = (df['comp_1'] + df['comp_2'] + df['comp_3']) / 3
df['Autonomy'] = (df['aut_1'] + df['aut_2'] + df['aut_3']) / 3
df['Relatedness'] = (df['rel_1'] + df['rel_2'] + df['rel_3']) / 3
df['Presence/Immersion'] = (df['pres_1'] + df['pres_2'] + df['pres_4'] + df['pres_5'] + df['pres_6'] + df['pres_7'] + df['pres_8'] + df['pres_9']) / 9
df['Intuitive Controls'] = (df['ctrl_1'] + df['ctrl_2'] + df['ctrl_3']) / 3
```
